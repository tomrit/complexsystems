```python
import numpy as np
import matplotlib.pyplot as plt
import shinriki
```

```python
t_max = 1
t_step = 1e-5
r0 = [0, 0.5, 0.75e-3]
dgl = shinriki.run(r0, t_max, t_step)
```

```python
start_idx=300
rt = np.array(dgl.rt[start_idx:])
```

```python
%matplotlib inline
plt.plot(rt[:,1])

zero_crossings1 = np.where(np.diff(np.sign(rt[:,1]))==-2)[0]
zero_crossings = shinriki.get_zero_crossings(rt[:,1])
plt.plot(zero_crossings, np.zeros(zero_crossings.size), 'or')
plt.plot(zero_crossings1, np.zeros(zero_crossings1.size), 'ob')
plt.xlim(0,4000)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
start_idx = 0

ax.plot(rt[start_idx:, 0], rt[start_idx:, 1], rt[start_idx:, 2])
ax.scatter(rt[zero_crossings,0], rt[zero_crossings,1], rt[zero_crossings,2], marker='.', color='r')
ax.set_xlabel('V_1 [V]')
ax.set_ylabel('V_2 [V]')
ax.set_zlabel('I_3 [A]')
```

```python
plt.plot(rt[zero_crossings,0], rt[zero_crossings,2], '.')
```
